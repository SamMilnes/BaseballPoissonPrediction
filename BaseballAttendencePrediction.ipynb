```python
import pandas as pd
from patsy import dmatrices
import numpy as np
import statsmodels.api as sm
```

```python
df = pd.read_csv('p3all_teams_full_season.csv', header=0)
df
```

```python
#Print Dataframe to csv

# df.to_csv('file_name.csv')
```

```python
#Print all games with NA values

null_data = df[df.isnull().any(axis=1)]
null_data
```

```python
# replace all na's with 0
df['STREAK'] = df['STREAK'].replace(np.nan, 0)
df
```

```python
# Remove all Na values
newdf = df.dropna()
newdf
```

```python
df_train = newdf
df_train
```

```python
expr = """LN_ATT ~ WIN_PCT  + CLI + STREAK + LN_HIGHEST_PSAL + LN_AVG_PSAL + LN_AGE + 
LN_CITY_POP + LN_STATE_INC + LN_CAP + STRIKE94_RECOV + NEW_PARK + POST_911"""
expr
```

```python
y_train, X_train = dmatrices(expr, df, return_type='dataframe')
```

```python
print(y_train)
```

```python
print(X_train)
```

```python
poisson_training_results = sm.GLM(y_train, X_train, family=sm.families.Poisson()).fit()
```

```python
print(poisson_training_results.summary())
```

```python
poisson_predictions = poisson_training_results.get_prediction(X_train)
predictions_summary_frame = poisson_predictions.summary_frame()
print(predictions_summary_frame)

predicted_counts=predictions_summary_frame['mean']
```

```python
pc = predicted_counts.to_frame()
pc
```

```python
df_train
```

```python
df_results = df_train
df_final = pd.merge(df_results, pc, left_index=True, right_index=True)
df_final
```

```python
col = df_final.pop('Predicted Attendance')
  
df_final.insert(4, 'Predicted Attendance', col)

df_final

```
